# 3.Which product sold the most?

```python
count_df = all_data.groupby(['Product']).agg({'Quantity Ordered':'sum' , 'Price Each':'mean'}) # Quantity Ordered is key and value is sum same for other.
```

```python
count_df
```

```python
count_df = count_df.reset_index()
```

```python
count_df
```

```python
products = count_df['Product'].values
```

```python
fig , ax1 = plt.subplots()

ax2 = ax1.twinx() ## as X-axis is same for both plots , ie we are sharing x-axis , ie we have to use ax2 = ax1.twinx()

ax1.bar(count_df['Product'] , count_df['Quantity Ordered'] , color='g')
ax2.plot(count_df['Product'] , count_df['Price Each'] )
ax1.set_xticklabels(products , rotation='vertical' , fontsize=10 )

ax1.set_ylabel('Order Count')
ax2.set_ylabel('Avg price of product')
```
